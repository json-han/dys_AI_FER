# Test 분류

```python
import matplotlib as plt
import pandas as pd

# CSV 불러오기
df = pd.read_csv("test_prediction_results.csv") 

# 실제 라벨별로 예측 결과 분포 구하기 (그룹화)
grouped = df.groupby(["true_label", "predicted_label"]).size().unstack(fill_value=0)
print("실제 라벨별 예측 결과 분포:")
grouped
```

```python
# ② 틀린 예측만 필터링
wrong_preds = df[df["true_label"] != df["predicted_label"]]

# 실제 라벨별로 틀린 예측의 분포
wrong_grouped = wrong_preds.groupby(["true_label", "predicted_label"]).size().unstack(fill_value=0)
print("\n틀린 예측 시, 예측된 라벨 분포:")
wrong_grouped
```

```python
# 필요 시 비율(%)로 변환
wrong_grouped_percent = wrong_grouped.div(wrong_grouped.sum(axis=1), axis=0) * 100
print("\n틀린 예측 분포(백분율):")
wrong_grouped_percent.round(1)
```

# Val 분류

```python
import pandas as pd

# CSV 불러오기
df = pd.read_csv("val_prediction_results.csv") 

# 실제 라벨별로 예측 결과 분포 구하기 (그룹화)
grouped = df.groupby(["true_label", "predicted_label"]).size().unstack(fill_value=0)
print("실제 라벨별 예측 결과 분포:")
print(grouped)

# ② 틀린 예측만 필터링
wrong_preds = df[df["true_label"] != df["predicted_label"]]

# 실제 라벨별로 틀린 예측의 분포
wrong_grouped = wrong_preds.groupby(["true_label", "predicted_label"]).size().unstack(fill_value=0)
print("\n틀린 예측 시, 예측된 라벨 분포:")
print(wrong_grouped)

# 필요 시 비율(%)로 변환
wrong_grouped_percent = wrong_grouped.div(wrong_grouped.sum(axis=1), axis=0) * 100
print("\n틀린 예측 분포(백분율):")
print(wrong_grouped_percent.round(1))
```
